```python
import pandas as pd
```

```python
pokemons = pd.read_csv('Pokemon.csv')
pokemons
```

```python
pokemons.columns
```

```python
documents = []
for _,row in pokemons.iterrows():
    doc = f"""
    Name: {row['name']}
    Generation: {row['generation']}
    Type 1: {row['type1']}
    Type 2: {row['type2']}
    Species: {row['species']}
    Height: {row['height']}
    Weight: {row['weight']}
    Ability 1: {row['ability1']}
    Ability 2: {row['ability2']}
    Hidden Ability: {row['hidden_ability']}
    Health Points(HP): {row['hp']}
    Attack: {row['attack']}
    Defence: {row['defense']}
    Special Attack: {row['sp_atk']}
    Special Defense: {row['sp_def']}
    Speed: {row['speed']}
    Total: {row['total']}
    EV Yield: {row['ev_yield']}
    Catch Rate: {row['catch_rate']}
    Base Friendship: {row['base_friendship']}
    Base Experience: {row['base_exp']}
    Growth Rate: {row['growth_rate']}
    Egg Group 1: {row['egg_group1']}
    Egg Group 2: {row['egg_group2']}
    Male Percentage: {row['percent_male']}
    Female Percentage: {row['percent_female']}
    Egg Cycles: {row['egg_cycles']}
    Special Group: {row['special_group']}
    """
    documents.append(doc)
```

```python
documents[0]
```

```python
with open("pokemon_documents.txt",'w') as f:
    for doc in documents:
        f.write(doc + "\n")
```
